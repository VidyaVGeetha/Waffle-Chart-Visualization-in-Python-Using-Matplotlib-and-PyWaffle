```python
# Waffle Charts 
#Step 11: # let's create a new dataframe for these three countries 
df_DNS=df_can.loc[['Denmark', 'Norway','Sweden'],:]
# let's take a look at our dataframe
df_DNS
```

```python
#Step 12: # compute the proportion of each category with respect to the total
total_values=df_DNS['Total'].sum()
category_proportions=df_DNS['Total']/total_values
#Step 13: Display Category Proportions as a Table
pd.DataFrame({"Category Proportion":category_proportions})
```

```python
#Step 13: The 13th step is defining the overall size of the waffle chart.
#Width of the chart
width=40
#Height of the chart
height=10
#Total number of Tiles
total_num_tiles=width*height
#print total number of Tiles
print(f'Total Number of Tiles is:{total_num_tiles}.')
```

```python
#Step 14:compute the number of tiles for each category
tiles_per_category=(category_proportions*total_num_tiles).round().astype(int)

#Step 15: # print out number of tiles per category
pd.DataFrame({'Number of tiles per Category':tiles_per_category})
```

```python
#  Step 16: Make the Waffle Chart Grid

#  16.1: Create an Empty Chart  
# Make a blank grid filled with zeros — this will later hold the numbers for each country.

waffle_chart = np.zeros((height, width), dtype=np.uint)

# 16.2: Start Counting Variables  

category_index = 0
tile_index = 0

# 16.3: Go Through Each Box  

for col in range(width):
    for row in range(height):
        tile_index += 1

        #  16.4 Check If It’s Time to Change Country  
        # If the current country’s share is already filled, move on to the next one.
        if tile_index > sum(tiles_per_category[0:category_index]):
            category_index += 1       

        #  16.5 Fill the Box with the Country Number  
        # Put the country’s number (0, 1, 2, etc.) into that box of the grid.
        waffle_chart[row, col] = category_index

#  16.6 Confirm Completion

print('Waffle chart populated!')
```

```python
#Step 18: Map the waffle chart matrix into a visual.
#18.1:  Create a Blank Canvas
   # (Make a new empty figure where the waffle chart will be drawn.)
fig=plt.figure()

# 18.2: Choose a Colour Style  
  #( Select the colour theme (blue to red) for displaying the waffle chart.)
colormap = plt.cm.coolwarm

# 18.3: Display the Waffle Chart  
# (Show the waffle_chart matrix as coloured blocks.)

plt.matshow(waffle_chart,cmap=colormap)

# 18.4 Add a Colour Bar  
# (Add a small scale on the side to show what each colour means.)
plt.colorbar()

#18.5  Display the finished waffle chart.
plt.show

```

```python
# Step 21: Define Function to Create a Waffle Chart from Categories and Values
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

def create_waffle_chart(categories, values, height, width, colormap, value_sign=''):
    # ---------------------------------------------
    # 21.0: Build the waffle_chart matrix first
    # ---------------------------------------------
    total_num_tiles = height * width
    values = np.array(values)
    proportions = values / values.sum()
    tiles_per_category = (proportions * total_num_tiles).round().astype(int)

    # initialize empty matrix
    waffle_chart = np.zeros((height, width), dtype=np.uint)
    category_index = 0
    tile_index = 0

    for col in range(width):
        for row in range(height):
            tile_index += 1
            if tile_index > tiles_per_category[:category_index].sum():
                category_index += 1
            waffle_chart[row, col] = category_index

    # ---------------------------------------------
    # 21.1: Create a Blank Canvas
    # ---------------------------------------------
    fig = plt.figure()

    # 21.2: Choose a Colour Style 
    # (use the colormap passed into the function)
    # colourmap = plt.cm.coolwarm   # old line
    colourmap = colormap

    # 21.3: Show the waffle_chart matrix as coloured blocks
    plt.matshow(waffle_chart, cmap=colourmap)

    # 21.4: Add a Colour Bar 
    plt.colorbar()

    # 21.5: get current axis
    ax = plt.gca()

    # 21.6: Set minor tick marks
    ax.set_xticks(np.arange(-.5, (width), 1), minor=True)
    ax.set_yticks(np.arange(-.5, (height), 1), minor=True)

    # 21.7: add gridlines based on minor ticks
    ax.grid(which='minor', color='w', linestyle='-', linewidth=2)

    # 21.8: Hide tick numbers
    plt.xticks([])
    plt.yticks([])

    # ---------------------------------------------
    # 21.9: Compute Cumulative and Final Total Values for Legend Generation
    # (use the values passed to the function, not df_DNS)
    # ---------------------------------------------
    values_cumsum = np.cumsum(values)
    total_values = values_cumsum[len(values_cumsum) - 1]

    # 21.10: Loop through each category and create Legend
    legend_handles = []
    for i, category in enumerate(categories):
        # Create Label String Combining Category Name and Total Value
        label_str = f"{category} ({values[i]}{value_sign})"

        # Map Category Value to Color Using Colormap Scale
        color_val = colormap(float(values_cumsum[i]) / total_values)

        # Add Color Patch and Label to Legend Handle List
        legend_handles.append(mpatches.Patch(color=color_val, label=label_str))

    # 21.11: Finally, show the legend
    plt.legend(
        handles=legend_handles,
        loc='lower center',
        ncol=len(categories),
        bbox_to_anchor=(0., -0.2, 0.95, .1)
    )

    plt.show()
```

```python
create_waffle_chart(
    categories=df_DNS.index.values,
    values=df_DNS['Total'].values,
    height=10,
    width=40,
    colormap=plt.cm.Set3,
    value_sign=''
)
```
